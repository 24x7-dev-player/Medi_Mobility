```python
import pandas as pd
import numpy as np
import random

# Function to generate random coordinates
def get_random_coordinates():
    return (random.uniform(0, 20), random.uniform(0, 20))

# Generate dummy data
num_patients = 100  # Adjust this number as needed
data = {
    'Patient ID': [],
    'Pickup Location': [],
    'Dropoff Location': [],
    'Appointment Time': [],
    'Medical Conditions': []
}

start_time = pd.Timestamp('2024-06-19 08:00')
end_time = pd.Timestamp('2024-06-19 18:00')

for _ in range(num_patients):
    data['Patient ID'].append(len(data['Patient ID']) + 1)
    data['Pickup Location'].append(get_random_coordinates())
    data['Dropoff Location'].append(get_random_coordinates())
    random_time = start_time + pd.Timedelta(minutes=random.randint(0, (end_time - start_time).seconds//60))
    data['Appointment Time'].append(random_time.time())
    data['Medical Conditions'].append(random.choice(['Heart Disease', 'Diabetes', 'Asthma']))

# Convert to DataFrame
df = pd.DataFrame(data)

df.to_csv("df_with_coordinates.csv", sep=',', index=False, encoding='utf-8')
```

```python
import pandas as pd
import random
from geopy.distance import geodesic

# Function to generate random coordinates within Gurugram (approximate bounding box)
def get_random_coordinates_within_gurugram():
    lat_min, lat_max = 28.4000, 28.5700
    lon_min, lon_max = 76.9500, 77.2000
    return (random.uniform(lat_min, lat_max), random.uniform(lon_min, lon_max))

# Generate 10 random dropoff locations within Gurugram
dropoff_locations = [get_random_coordinates_within_gurugram() for _ in range(10)]

# Generate dummy data
num_patients = 100  # Adjust this number as needed
data = {
    'Patient ID': [],
    'Pickup Location': [],
    'Dropoff Location': [],
    'Appointment Time': [],
    'Medical Conditions': []
}

# Define the start and end time for appointments
start_time = pd.Timestamp('2024-06-19 10:00')
end_time = pd.Timestamp('2024-06-19 19:00')

# Generate possible appointment times with a 30-minute gap
appointment_times = pd.date_range(start_time, end_time, freq='30T').time

for _ in range(num_patients):
    data['Patient ID'].append(len(data['Patient ID']) + 1)
    data['Pickup Location'].append(get_random_coordinates_within_gurugram())
    data['Dropoff Location'].append(random.choice(dropoff_locations))  # Assign from the 10 pre-generated coordinates
    random_time = random.choice(appointment_times)
    data['Appointment Time'].append(random_time)
    data['Medical Conditions'].append(random.choice(['Heart Disease', 'Diabetes', 'Asthma']))

# Convert to DataFrame
df = pd.DataFrame(data)

# Display the first few rows of the dataframe
df.head()
```

```python
df.to_csv("df_like_real.csv", sep=',', index=False, encoding='utf-8')
```
